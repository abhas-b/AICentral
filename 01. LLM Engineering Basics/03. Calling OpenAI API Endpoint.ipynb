```python
import requests
from dotenv import load_dotenv
import os
```

```python
load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")

header = {"Authorization":f"Bearer {api_key}", "Content-Type": "application/json"}

payload = {
    "model":"gpt-4o-mini",
    "messages":[
        {"role":"user", "content":"Hi GPT. Taking you for a test drive"}
    ]
}

response = requests.post(
    url="https://api.openai.com/v1/chat/completions",
    headers=header,
    json=payload
)

response.status_code
```

```python
response.json()["choices"][0]["message"]["content"]
```

# Use OpenAI API for Groq

```python
from openai import OpenAI


groq_url = "https://api.groq.com/openai/v1"
groq_api_key = os.getenv('GROQ_API_KEY')
groq = OpenAI(api_key=groq_api_key, base_url=groq_url)
```

```python
tell_a_fact = [
    {"role":"user", "content":"Tell me a fun fact"}
]

groq_model = "llama-3.1-8b-instant"
response = groq.chat.completions.create(model=groq_model, messages=tell_a_fact)
response.choices[0].message.content
```
