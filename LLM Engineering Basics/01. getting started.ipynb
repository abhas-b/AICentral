```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Image, display, Markdown
from scraper import fetch_website_contents
from utilities import usage_metrics
```

```python
load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    raise ValueError("OPENAI_API_KEY is not set")
elif api_key.startswith("sk-proj"):
    print("Valid OPENAI API key found")
```

```python
message = "Hi GPT. This is my first message to you. How can you help me generating project ideas for Generative AI and Agentic AI?"

messages = [
    {
        "role":"user",
        "content":message
    }
]

messages
```

```python
openai = OpenAI(api_key=api_key)
response = openai.chat.completions.create(model="gpt-4o-mini", temperature=0.2, messages=messages)
```

```python
response.choices[0].message.content[:100]
```

```python
usage_metrics(response=response)
```

```python
system_prompt = """You are a helpful assistant. You are able to look webpages from a company listing and extract and summarize key information about the company.
Ignore irrelevant information. 
Respond in Markdown"""

user_prompt_prefix = """Here are the contents of the website. Provide a short summary of the key information presented on the page in markdown. 
If it provides pricing information or any announcements or discounts, capture that too. """
```

```python
prompt = [
    {"role":"system", "content":system_prompt},
    {"role":"user", "content":user_prompt_prefix}
]
```

```python
response = openai.chat.completions.create(model='gpt-4.1-nano', messages=prompt)
usage_metrics(response)
```
