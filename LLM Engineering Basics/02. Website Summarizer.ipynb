```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Image, display, Markdown
from scraper import fetch_website_contents
from utilities import usage_metrics
```

```python
load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")

if not api_key:
    raise ValueError("OPENAI_API_KEY is not set")
elif api_key.startswith("sk-proj"):
    print("Valid OPENAI API key found")
```

```python
system_prompt = """You are a helpful assistant. You are able to look webpages from a company listing and extract and summarize key information about the company.
Ignore irrelevant information. 
Respond in Markdown. Do not wrap the markdown in code block - respond just with a markdown."""

user_prompt_prefix = """Here are the contents of the website. Provide a short summary of the key information presented on the page in markdown. 
If it provides pricing information or any announcements or discounts, capture that too. 

Website content: 



"""
```

```python
def create_messages(website_content):
    prompt = [
    {"role":"system", "content":system_prompt},
    {"role":"user", "content":user_prompt_prefix+website_content}
    ]
    return prompt
```

```python
openai = OpenAI(api_key=api_key)
```

```python
def summarize(url):
    page = fetch_website_contents(url)
    messages = create_messages(page)
    response = openai.chat.completions.create(model='gpt-4.1-nano', messages=messages)
    print("Token summary: ", usage_metrics(response=response))
    return response.choices[0].message.content

url = "https://www.indiamart.com/proddetail/adhesive-cable-clip-silicone-cord-organizer-for-appliances-27478227955/"
response = summarize(url=url)
```
