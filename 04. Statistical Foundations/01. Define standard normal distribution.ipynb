```python
from scipy import stats
import matplotlib.pyplot as plt
from scipy.stats import norm
import numpy as np
```

```python
normal_dist = stats.norm(loc=0, scale=1)
```

```python
pdf_at_zero = normal_dist.pdf(0)
print(f"Probability density at 0: {pdf_at_zero}")
```

```python
# Generate 5 random samples

normal_dist.rvs(size=5)
```

```python
# Calculate cumulative probability P(X<=1)

normal_dist.cdf(1)
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, normal_dist.pdf(x))
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.show()
```
