```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
df.head()
```

```python
# event A: Passenger survived (P(A))

p_a = np.sum(df['Survived'] == 1)/df['PassengerId'].nunique()

print(f"Survival Probability: {p_a:1%}")
```

```python
# event B: Passenger is female: (P(B))

p_b = np.sum(df['Sex'] == "female")/df['PassengerId'].nunique()
print(f"Female Probability: {p_b:1%}")
```

```python
# P(A and B)

p_a_and_b = df.loc[(df['Sex'] == 'female') & (df['Survived'] == 1), 'PassengerId'].nunique()/df['PassengerId'].nunique()
```

```python
# Probability of survival given female: P(A|B)

p_a_given_b = p_a_and_b/p_b
```

```python
if abs(p_a_given_b-p_a) < 1e-6:
    print("independent events")
else:
    print("dependent events")
```
